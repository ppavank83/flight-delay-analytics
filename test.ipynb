```python
import pandas as pd
import numpy as np

# Load the CSV
df = pd.read_csv("data/T_ONTIME_REPORTING.csv")

# Convert FL_DATE to datetime.date
df["FL_DATE"] = pd.to_datetime(df["FL_DATE"], errors='coerce').dt.date

# Replace NaN with None for SQL compatibility
df = df.replace({np.nan: None})

# Optional: Validate number of columns matches your SQL insert
print("âœ… Sample cleaned row with expected 25 values:", len(df.iloc[0]))
```
